### Line plot

```python
x_val=[0,1,2]
y_val=[3,5,9]
plt.plot(x_val,y_val)
```

### Bar plot

```python
cat_counts=data['Genre'].value_counts()# frequency of each genre|
cat_counts
```

Visualization of frequency of genre using bar chart

```python
x_bar=cat_counts.index
x_bar
```

```python
y_bar=cat_counts.values# or y_bar=cat_counts Both result the same
y_bar
```

```python
plt.bar(x_bar,y_bar)
```

```python
# to change the size of the figure
plt.figure(figsize=(12,8))
plt.bar(x_bar,y_bar)
```

Rotate the tick labels and also increse the font size

```python
plt.figure(figsize=(12,8))
plt.xticks(rotation=45,fontsize=12)
plt.bar(x_bar,y_bar)
```

```python
# We can also change the width and colors of the bars
plt.figure(figsize=(12,8))
plt.xticks(rotation=45,fontsize=10)
plt.bar(x_bar,y_bar,width=0.3,color='orange')
```

```python
plt.figure(figsize=(12,8))
plt.xticks(rotation=45,fontsize=10)
plt.bar(x_bar,y_bar,width=0.3,color='orange')
plt.title('Frequency of Genre',fontsize=10)
plt.xlabel('Genre',fontsize=10)
plt.ylabel('Frequency',fontsize=10)
```

```python
sns.countplot(x='Genre',data=data,order=data['Genre'].value_counts().index)
plt.xticks(rotation=45)
```

### Pie chart

```python
sales_data = data[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
region_sales = sales_data.T.sum(axis='columns')

plt.pie(region_sales,
       labels=region_sales.index,
       startangle=90,
       explode=(0.2,0,0,0),
       autopct = '%.2f%%') # label the wedges with their numeric value

plt.title('Total Sales across various Regions')
plt.show()
```

### Histogram

```python
plt.hist(data['Year'])
plt.show()
```

```python
plt.hist(data['Year'],bins=20)# default bins =10
plt.show()
```

```python
sns.histplot(data=data['Year'],bins=10)# Using seaboen
plt.show()
```

### KDE plot(kernel Density Estimate)

```python
sns.kdeplot(data=data['Year'])
plt.show()
```

### Box Plot

```python
# this is also called box and whisker plot
plt.figure(figsize=(15,10))
sns.boxplot(y = data["Global_Sales"])
plt.yticks(fontsize=20)
plt.ylabel('Global Sales (in million dollars)', fontsize=20)
plt.title('Global Sales of video games', fontsize=20)
```

```python
sns.histplot(data['Global_Sales'],bins=10)
```
